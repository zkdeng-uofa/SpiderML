```python
def map_taxon_ids(taxon_ids, lookup_dict):
    result = {'phylum': None, 'class': None, 'order': None, 'family': None, 'genus': None}
    for taxon_id in taxon_ids:
        taxon_info = lookup_dict.get(int(taxon_id))
        if taxon_info:
            rank = taxon_info['rank']
            if rank in result:
                result[rank] = taxon_info['name']
    return result
```

```python
test[['phylum', 'class', 'order', 'family', 'genus']] = test['taxon_ids'].apply(lambda ids: pd.Series(map_taxon_ids(ids, taxonomy_dict)))
test
```
